```python
# data_ebit_ttm = data.quarterly_financials['EBIT'].sum(axis=1)
# data_ebit_ttm
test = data.quarterly_financials
list_col = test.columns
test = test.reset_index()
print(list_col)
test['ttm'] = 0
if len(list_col) >=4 :
    ttm_period = 4
else : ttm_period = len(list_col)
for i in range(ttm_period): 
    test['ttm'] = test['ttm'] + test[list_col[i]]

test = test.set_index('index')
test
```
